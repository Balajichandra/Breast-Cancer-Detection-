```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
data.head(5)
```

```python
data.columns
```

```python
Y = data.diagnosis
list = ['Unnamed: 32','id','diagnosis']
X = data.drop(list,axis=1)
```

```python
ax = sns.countplot(Y,label="Count")
B,M = Y.value_counts()
print("No of Begining:",B)
print("No of Maligant:",M)
```

```python
X.describe()
```

```python
X.info()
```

```python
drop_list = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
X_new = X.drop(drop_list,axis=1)
X_new
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_new,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(xtrain,ytrain)
```

```python
import pickle
filename = 'brest-cancer-prediction-model.pkl'
pickle.dump(model, open(filename, 'wb'))
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
print("Accuracy Score:",accuracy_score(ytest,model.predict(xtest)))
print("Confusion matrix:","\n",confusion_matrix(ytest,model.predict(xtest)))
```
